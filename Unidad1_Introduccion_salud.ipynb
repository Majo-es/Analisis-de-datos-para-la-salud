### 1. ¿Por qué es importante el análisis de datos en el ámbito de la salud? <a name="id1"></a>

#### 1.1 Introducción

El análisis de datos desempeña un papel fundamental en todos los ámbitos del campo de la salud, desde la investigación médica hasta la atención al paciente y la gestión de la salud pública, ya que permite extraer información significativa y tomar decisiones basadas en la evidencia.

Toda esta información recogida por los profesionales sanitarios y los dispositivos médicos y los datos obtenidos y derivados de estudios clínicos y de investigación pueden ser analizados y usados para mejorar la atención a los pacientes, entender y tratar más eficientemente las enfermedades y ofrecer una medicina personalizada.

El sector de la salud se caracteriza por generar una cantidad enorme de datos relacionados con pacientes, enfermedades, centros hospitalarios y de salud, resultados de investigación, etc. Además, tener distintas fuentes conlleva a la presencia de datos de muchos tipos, por ejemplo, datos clínicos o analíticos de pacientes, imágenes de pruebas médicas, datos genómicos o epidemiológicos, señales biomédicas, entre otros.

Es bien sabido que, cuantos más datos, más robusto es un análisis, ya que las muestras son más representativas del conjunto real y su variabilidad. Sin embargo, la calidad de los datos debe ser buena y, para ello, debe haber un control y preproceso de los datos y sus fuentes.

Así pues, lidiar con tanta información, muchas veces desestructurada y heterogénea, supone también un reto. En primer lugar, los datos de salud requieren un alto nivel de privacidad, de modo que es necesario tener infrastructuras y procesos que garanticen el buen tratamiento y anonimidad de los datos. Por otro lado, al haber distintos tipos y fuentes de datos, el tratamiento y análisis de estos es muy diverso y requiere normalmente bastante experiencia. Por lo tanto, es esencial formar especialistas para que puedan trabajar con datos de estas características y en un sector tan cambiante y diverso.

Si ponemos de ejemplo un hospital, cada segundo se genera información, pero solo una pequeña parte se recoge, y, de esta parte, no toda se procesa de manera correcta. Actualmente, la limitación del análisis de datos en salud, no viene por la complejidad técnica de los datos o los modelos, sino por la infrastructura que se necesita detrás.

En este sentido, en los últimos años ha habido un gran esfuerzo enfocado en la digitalización, que permita tener la información estructurada y organizada para poder acceder a la información de manera más rápida y eficaz. Este paso es crítico para poder aprovechar todas las ventajas que el análisis de datos puede ofrecer. Es necesario destacar también que la información médica se puede representar con distintos formatos (datos tabulares, temporales, imágenes, audio, video, etc), de modo que es necesario tener plataformas que permitan almacenar grandes volúmenes de datos, y de distinto tipo.

#### 1.2 Evolución del análisis de datos en salud

El análisis de datos en el ámbito de la salud ha experimentado una evolución significativa en las últimas décadas, impulsada por avances tecnológicos, la disponibilidad de grandes volúmenes de datos y la creciente necesidad de tomar decisiones basadas en evidencia.

En las etapas iniciales, el análisis de datos en salud se centraba principalmente en **describir características** básicas de la población, como la edad, el sexo, el diagnóstico y los tratamientos recibidos. Los informes estadísticos y los resúmenes descriptivos eran comunes en este enfoque.

Con el desarrollo de técnicas estadísticas más avanzadas, el **análisis inferencial** ganó importancia. Este enfoque permitió a los investigadores realizar inferencias sobre la población general basadas en muestras de datos, utilizando métodos como la regresión, el análisis de varianza y las pruebas de hipótesis.

La revolución de los algoritmos de aprendizaje y las mejoras tecnológicas que permitieron trabajar con grandes conjuntos de datos también tuvieron un gran impacto en este campo. En este contexto, el **análisis predictivo** se convirtió en una herramienta poderosa en el ámbito de la salud, ya que los modelos predictivos pueden predecir resultados clínicos, identificar factores de riesgo y personalizar tratamientos para pacientes individuales. En esta época también ganó importancia el **análisis prescriptivo**, que
implica el uso de análisis avanzados para recomendar acciones específicas basadas en los resultados del análisis predictivo. Por ejemplo, los sistemas de apoyo a la decisión clínica pueden utilizar datos para recomendar tratamientos óptimos o intervenciones preventivas para mejorar los resultados de salud.

Actualmente, con el avance de la tecnología de la información y la comunicación, el análisis de datos en salud se ha vuelto más ágil y dinámico. Los **sistemas de monitoreo en tiempo real** pueden detectar tendencias emergentes, brotes de enfermedades y eventos adversos de manera oportuna, permitiendo una respuesta rápida y efectiva. También es importante destacar que, con la diversificación de fuentes de datos en salud, como registros médicos electrónicos, datos de dispositivos portátiles y datos genómicos, se ha vuelto crucial integrar y analizar estos datos de manera conjunta para obtener una comprensión completa de la salud y el bienestar de los pacientes.

#### 1.3 Aplicaciones de análisis de datos en salud

El objetivo del análisis de datos en el ámbito sanitario es extraer información relevante de los datos para mejorar la atención y el tratamiento médico, prevenir o detectar precozmente enfermedades y optimizar los sistemas de salud. En este contexto, algunas de las aplicaciones más comunes son:

**Medicina personalizada**
- Seguimiento y análisis de datos de pacientes con enfermedades crónicas para mejorar la gestión de la atención.
- Desarrollo de intervenciones personalizadas basadas en datos para mejorar los resultados de salud.

**Investigación clínica y desarrollo de medicamentos**
- Análisis de datos de ensayos clínicos para evaluar la eficacia y seguridad de nuevos tratamientos.
- Identificación de subgrupos de pacientes que podrían beneficiarse más de ciertos tratamientos.
- Análisis de datos genómicos para comprender mejor las bases moleculares de las enfermedades.

**Análisis de datos de dispositivos médicos conectados**
- Monitoreo de pacientes a través de dispositivos portátiles para prevenir enfermedades y mejorar la salud.
- Análisis de datos de monitoreo remoto para la gestión de enfermedades crónicas y la toma de decisiones clínicas.

**Monitoreo de epidemias y brotes**
- Utilización de datos epidemiológicos para identificar patrones de propagación de enfermedades.
- Detección temprana de brotes y respuesta rápida para limitar la propagación de enfermedades infecciosas.

**Optimización de recursos hospitalarios**
- Análisis de datos de admisiones, tiempos de espera y ocupación de camas para mejorar la planificación de recursos.
- Identificación de áreas de mejora en la eficiencia operativa de los hospitales.
- Análisis de factores de riesgo para prever y prevenir readmisiones hospitalarias.


**Análisis de series temporales para predicción y planificación**
- Uso de modelos de series temporales para prever la demanda futura de servicios de salud.
- Planificación de recursos y personal basada en las predicciones de carga de trabajo.

### 2. Tipos de datos en salud <a name="id2"></a>

Los datos en salud abarcan una amplia variedad de información relacionada con la salud de individuos, centros e instituciones sanitarias y poblaciones. Por este motivo, tienen tipos muy diversos.

**Datos tabulares**

- Datos demográficos: incluyen información sobre la edad, género, raza, etnia y otros factores demográficos de los individuos.

- Datos clínicos: contienen información específica sobre la salud de un individuo, como diagnósticos médicos, procedimientos, tratamientos, historial de medicamentos y resultados de pruebas de laboratorio.

- Datos de ensayos clínicos: información recopilada durante la realización de estudios clínicos como protocolos de investigación, resultados y datos demográficos de los participantes.

- Datos de prescripción y farmacia: detalles sobre medicamentos recetados, incluyendo dosis, frecuencia y duración del tratamiento.

- Datos de salud mental y estilo de vida: incluyen información sobre el estado emocional y mental de un individuo, y sobre hábitos y comportamientos relacionados con la salud.

- Datos de administración de la salud: incluyen información utilizada en la gestión de instituciones de salud, como datos financieros, programación de citas y asignación de recursos.

**Series temporales**
- Datos de dispositivos médicos conectados: información generada por dispositivos de monitoreo de salud, como monitores de glucosa, monitores de presión arterial y dispositivos de seguimiento de actividad física.

- Datos de vigilancia epidemiológica: datos usados para monitorear la incidencia y prevalencia de enfermedades en poblaciones, incluyendo brotas y tendencias a lo largo del tiempo.

**Texto**
- Datos de historias clínicas electrónicas: registros electrónicos de la información médica de un paciente, que pueden incluir notas de médicos, resultados de exámenes y otros detalles relevantes.

**Imágenes**

- Datos de imágenes médicas: incluyen radiografías, resonancias magnéticas, tomografías computarizadas y otras imágenes utilizadas en el diagnóstico y seguimiento de enfermedades.

**Información genética (Mixto)**
- Datos genómicos: información sobre la secuencia genética de un individuo, incluyendo variantes genéticas, mutaciones y otros detalles relacionados con el ADN. Incluyen distintos tipos de datos (tabular, texto, imágenes).

### 3. Proyectos y casos de éxito <a name="id3"></a>

Con el auge del análisis de datos y el Big Data en salud, numerosas empresas e instituciones han implementado proyectos que han tenido un impacto significativo en la mejora de la atención médica, la toma de decisiones y la investigación.

Algunos de los proyectos más destacados son:

**IBM Watson for Oncology**
- IBM Watson for Oncology utiliza inteligencia artificial para analizar grandes cantidades de literatura médica, datos clínicos y evidencia científica.
- Ayuda a los oncólogos a tomar decisiones informadas sobre opciones de tratamiento personalizadas para pacientes con cáncer.
- Se ha implementado en varios hospitales en todo el mundo, demostrando mejoras en la precisión de las recomendaciones de tratamiento.

**Project DataSphere**
- Project DataSphere es una iniciativa que permite compartir y analizar datos de ensayos clínicos oncológicos de manera colaborativa.
- Facilita el análisis conjunto de datos para descubrir patrones y tendencias que podrían llevar a avances en la investigación del cáncer.

**Proyecto MIMIC (Medical Information Mart for Intensive Care)**
- MIMIC es una base de datos que recopila información detallada de pacientes en unidades de cuidados intensivos (UCI).
- Ha sido fundamental para la investigación en áreas como la predicción de sepsis, identificación de factores de riesgo y mejora de la atención en entornos de cuidados intensivos.

**Proyecto Nightingale de Google y Ascension:**
- Google ha trabajado con el sistema de salud Ascension en el Proyecto Nightingale para analizar datos de pacientes y mejorar la atención médica.
- El proyecto tiene como objetivo utilizar la inteligencia artificial para identificar patrones y proporcionar información útil para la atención al paciente.

**Google's DeepMind Health**
- DeepMind (actualmente parte de Google) ha trabajado en colaboración con el Servicio Nacional de Salud (NHS) del Reino Unido.
- Utiliza algoritmos de aprendizaje profundo para analizar datos médicos, mejorando la eficiencia en la identificación y tratamiento de enfermedades oculares, así como en la gestión de la insuficiencia renal.
- Proyecto estrella: **AlphaFold**.
- AlphaFold tiene como objetivo predecir la estructura tridimensional de las proteínas a partir de su secuencia de aminoácidos.
- Utiliza redes neuronales profundas y técnicas de aprendizaje profundo (Deep Learning) para analizar patrones complejos en los datos y prever la forma en que una cadena de aminoácidos se pliega para formar una estructura tridimensional específica.




En España también hay numerosas startups y proyectos basados en análisis de datos en salud:

**Iomed**

- Iomed usa procesamiento de lenguaje natural (NLP) para transformar las historias clínicas electrónicas en datos estructurados, optimizando, así, la gestión de datos clínicos.
- A través del software desarrollado por la empresa, los hospitales pueden tener en una misma base de datos una gran cantidad de variables clínicas de pacientes anonimizados, con lo que consiguen reducir el tiempo empleado en el reclutamiento de los pacientes y, a su vez, acelerar la investigación clínica.

**Exheus**
- Spin-off del Institut de Recerca del Hospital Sant Pau, de la UPC y Summit SL.
- Usa técnicas de inteligencia artificial para analiza la expresión génica de tu cuerpo (ARNm) a partir de una pequeña muestra de sangre. Permite medir los niveles de expresión de los genes asociados a la salud metabólica, el estado de los sistemas del organismo, la longevidad y el estado inflamatorio.

**Sycai**
- Sycai es una empresa de desarrollo de software de inteligencia artificial  especializada en la aplicación de Deep Learning a la interpretación de imagen médica. El objetivo del programa es realizar un diagnóstico precoz y no invasivo de cualquier cáncer abdominal.
- Sycai Medical es una herramienta capaz de detectar, caracterizar y clasificar lesiones en los 5 subtipos de lesión pancreática existentes, con hasta un 90% de precisión.

**Idoven**
- Idoven combina algoritmos de inteligencia artificial en la nube con tecnología wearable para detectar de forma temprana problemas cardíacos (como alteraciones de la actividad eléctrica cardíaca y arritmias) y lograr enfermedades como el infarto de miocardio y la muerte súbita a través del análisis de electrocardiogramas de larga duración.
- El usuario debe monitorizar su corazón durante días o semanas con un dispositivo no invasivo.

### 4. Bases de datos <a name="id4"></a>

Uno de los grandes éxitos de la era digital es la creación de bases de datos de acceso libre que contienen datos de salud. Esta información es muy valiosa para la investigación y los análisis en este campo y ha permitido desarrollar proyectos e iniciativas muy importantes en el ámbito sanitario.

Actualmente hay muchos recursos disponibles con datos de salud, tanto bases de datos de distintos tipos, como repositorios de artículos e información de ensayos clínicos.

- [WHO Global Health Observatory](https://www.who.int/data/gho): proporciona una amplia variedad de datos relacionados con la salud mundial, incluyendo estadísticas sobre enfermedades, sistemas de salud y determinantes de la salud.

- [PubMed Central (PMC)](https://www.ncbi.nlm.nih.gov/pmc/): ofrece acceso gratuito a artículos científicos en el campo de la biomedicina y la salud.

- [OECD Health Statistics](https://stats.oecd.org/Index.aspx?DataSetCode=HEALTH_STAT): proporciona estadísticas comparativas sobre sistemas de salud y resultados en los países miembros de la Organización para la Cooperación y el Desarrollo Económicos (OCDE).

- [Healthdata.gov](https://www.healthdata.gov/): plataforma que proporciona acceso a varios conjuntos de datos relacionados con la salud en los Estados Unidos.

- [Instituto Nacional de Estadística](https://www.ine.es/index.htm): proporciona acceso a una amplia variedad de estadísticas y temas como población, empleo, comercio, turismo, precios, y otros indicadores económicos y sociales.

- [Medical ImageNet](https://aimi.stanford.edu/shared-datasets): esta base de datos se centra en imágenes médicas y ofrece un conjunto de datos específico para tareas de clasificación y segmentación en el ámbito médico.

- [Brain Image Library](https://www.brainimagelibrary.org/): BIL contiene imágenes cerebrales, como resonancias magnéticas estructurales y funcionales, utilizadas en investigaciones neurológicas.

- [GenBank](https://www.ncbi.nlm.nih.gov/genbank/): GenBank es una base de datos internacional de secuencias genéticas gestionada por el Centro Nacional de Información Biotecnológica (NCBI) en Estados Unidos. Contiene secuencias de ADN, ARN y proteínas, y es ampliamente utilizado en todo el mundo.

- [GWAS Catalog](https://www.ebi.ac.uk/gwas/): esta base de datos compila resultados de estudios de asociación de todo el genoma que buscan identificar variantes genéticas asociadas con enfermedades y rasgos complejos.
