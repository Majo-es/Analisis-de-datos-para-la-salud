### 1. Modelos de aprendizaje automático en el ámbito biomédico <a name="id1"></a>

#### 1.1 Introducción

El aprendizaje automático (en inglés, *machine learning*) en salud es un campo en rápido crecimiento que tiene como objetivo proporcionar analisis más complejos y eficientes para extraer información y mejorar la atención médica. Se puede dividir en dos categorías principales: aprendizaje supervisado y aprendizaje no supervisado.

```python
from IPython import display
display.Image("img/superv_unsuper.jpeg")
```

**Aprendizaje supervisado**

En el aprendizaje supervisado, el modelo se entrena utilizando un conjunto de datos que incluye ejemplos de entrada junto con las salidas deseadas correspondientes (etiquetas). El objetivo es que el modelo aprenda a mapear las entradas a las salidas correctas. Después de entrenar el modelo, se puede utilizar para hacer predicciones sobre nuevas instancias de datos. Este tipo de modelos incluye tanto problemas de clasificación, en el cual se quiere predecir correctamente la categoría/etiqueta de los datos, como de regresión, que tiene como objetivo predecir un valor numérico continuo.

```python
display.Image("img/supervised_ML.jpeg")
```

En el ámbito médico, los algoritmos supervisados tienen multitud de aplicaciones, entre las cuales podemos destacar las siguientes:

- Diagnóstico médico: usar distintos tipos de datos (por ejemplo imágenes médicas), junto con etiquetas que indican la presencia o ausencia de enfermedades, para entrenar un modelo de clasificación. Posteriormente, este modelo se puede puede usar con datos de pacientes nuevos para acelerar el diagnóstico.

- Pronóstico de enfermedades: predecir la progresión de una enfermedad o la probabilidad de recaída en pacientes con base en datos clínicos y biomédicos.

- Predicción de riesgos y complicaciones: estimar el riesgo de complicaciones postoperatorias para ayudar en la toma de decisiones clínicas y la planificación del tratamiento.

- Personalización de tratamientos: utilizar datos genómicos y clínicos para personalizar los regímenes de tratamiento y prever respuestas a terapias específicas.

- Detección de fraudes en seguros de salud: identificar posibles fraudes en reclamaciones de seguros médicos.

**Aprendizaje no supervisado**

En el aprendizaje no supervisado, el modelo se entrena en un conjunto de datos que no contiene etiquetas. El objetivo principal es descubrir patrones, estructuras o relaciones intrínsecas en los datos sin conocer de antemano las salidas esperadas. Este tipo de modelo es útil cuando se busca explorar la estructura subyacente de los datos.

```python
display.Image("img/unsupervised_ML.jpeg")
```

Algunos ejemplos de aplicaciones de modelos no supervisados en salud son:

- Agrupar pacientes: usar técnicas de agrupamiento para clasificar a los pacientes en grupos según sus perfiles clínicos, lo que puede ser útil para personalizar tratamientos y estrategias de atención.

- Descubrimiento de subtipos de enfermedades: identificar subgrupos de pacientes con características clínicas similares que podrían indicar variantes de una enfermedad específica, lo que puede llevar a tratamientos más personalizados.

- Segmentación de imágenes médicas: usar técnicas de segmentación para dividir imágenes médicas en regiones anatómicas específicas o identificar áreas con características particulares, lo que facilita el análisis clínico. También puede ser un método supervisado.


- Agrupar texto médico: aplicar técnicas de procesamiento de lenguaje natural (NLP) para agrupar documentos médicos, como historias clínicas, con el objetivo de identificar tendencias y patrones en la información no estructurada.

- Detección de anomalías: identificar patrones inusuales o datos atípicos que podrían indicar condiciones médicas raras, errores en los datos o eventos inesperados.

### 2. Modelos de aprendizaje automático supervisado <a name="id3"></a>

En el contexto del análisis supervisado, se pueden diferenciar los modelos usados para resolver problemas de clasificación, y los modelos modelos usados en problemas de regresión. En el primer caso, se hacen predicciones sobre una variable categórica (por ejemplo, tipo de cáncer, supervivencia de pacientes) mientras que en el segundo caso, la etiqueta es una variable numérica (por ejemplo, número de personas infectadas, precio de un tratamiento).

La mayoría de algoritmos presentan diferentes versiones y pueden aplicarse en los dos casos. Por ejemplo, la regresión lineal tiene su forma común para resolver regresiones, pero también cuenta con una versión para los problemas de clasificación, la regresión logística. Lo mismo sucede con los *support vector machine (SVM)*, que pueden aplicarse para regresión (*SV Regression*) y clasificación (*SV Classifier*). Los modelos basados en arboles (*Decision Tree*, *Random Forest*, *Ada Boost* y *XGBoost*) también pueden usarse para resolver ambos problemas.

A continuación, vamos a comentar algunos de los modelos más usados en el ámbito médico.

```python
display.Image("img/supervised_models.png")
```

**Modelo lineal**

Los modelos lineales son uno de los algoritmos más populares. Básicamente, un modelo lineal clásico representa la suma ponderada de las características de entrada más un elemento de intercepción o sesgo. El peso que tiene cada variable independiente nos da una idea de su contribución en relación a la variable dependiente. Dada su simplicidad, son ideales para explicar de manera más formal el concepto de modelo o hipótesis.

Los modelos lineales tradicionales se representan en un espacio de 2D, pero estos algoritmos pueden también aplicarse en N dimensiones. En este contexto N-dimensional, los *support vector machines* son unos de los métodos más usados. La limitación de los modelos lineales es que muchas veces la relación entre la variable que queremos predecir y las variables predictoras no es lineal, de modo que los modelos lineales pierden capacidad predictiva porque no son capaces de explicar todas las relaciones.



```python
display.Image("img/lineal_models_example.jpeg")
```

Por otro lado, los modelos lineales también se pueden aplicar a problemas de clasificación, siendo la regresión logística el modelo más popular en el ámbito biomédico. Se usa comunmente para predecir si una situación/enfermedad se va a producir o no. La regresión logística se considera un modelo lineal en el sentido de que busca una relación lineal entre las variables predictoras y la variable de respuesta transformada mediante una función logarítmica. Aunque la relación entre las variables predictoras y la probabilidad que ocurra el evento (e.g. enfermedad) no es lineal, la transformación logarítmica del cociente de probabilidades permite que la regresión logística se modele como un modelo lineal generalizado (GLM), una extensión del modelo lineal clásico.

En la regresión logística se obtiene la probabilidad de pertenecer a una clase, por ejemplo, la probabilidad de un paciente de padecer la enfermedad en función de las características introducidas en el modelo. A partir de esta probabilidad se puede establecer un punto de corte, que dará la clasificación de cada paciente.

```python
display.Image("img/Feature-importances-LR.png")
```

**K-Nearest Neighbors**


El algoritmo de k-NN (*K-Nearest Neighbors*) asume que objetos similares están cerca entre sí. Basándose en las distancias euclidianas entre muestras (u otras métricas de distancia más sofisticadas), el algoritmo de k-NN asigna iterativamente muestras a grupos, de manera que las muestras con valores de características similares se separan por distancias más cortas y, por lo tanto, se asignan al mismo grupo. A medida que se agregan muestras próximas a un grupo con el centroide más cercano, los centroides se recalculan. El objetivo final es minimizar la inercia, que es la suma de las distancias entre el centroide y las muestras dentro de un grupo

Sin embargo, es posible que las muestras agrupadas juntas pertenezcan a diferentes clases (por ejemplo, muestras de tumores benignos y malignos podrían estar en el mismo grupo). En esos casos, las muestras recién agregadas se asignan a la clase mayoritaria, a través de un proceso de votación (por ejemplo, si la mayoría de las muestras en ese grupo pertenecen a la clase de cáncer maligno, cualquier muestra benigna asignada erróneamente a ese grupo también se clasificará como maligna).

En el ámbito médico, este modelo se usa en problemas de clasificación, veamos un ejemplo del artículo `An Introduction to Machine Learning Approaches for Biomedical Research`. En este estudio se usó un modelo de k-NN para identificar la supervivencia de pacientes con hepatitis. El dataset contiene información clínica y de progresión de la enfermedad de 155 pacientes, 32 de los cuales no sobrevivieron, y 123, si. El 82% de pacientes se clasificaron correctamente según su supervivencia usando esta técnica.

**Modelos basados en árboles**

Los modelos de árbol de decisión se basan en una estructura de árbol donde cada nodo interno representa una característica (o atributo) y cada borde (rama) representa una decisión basada en esa característica. Los nodos hoja representan las etiquetas o valores de predicción. El algoritmo selecciona la característica que mejor divide los datos en subconjuntos más puros o homogéneos. Esto se hace maximizando una medida de pureza, como la ganancia de información, la reducción de la impureza de Gini o la entropía.  Los datos se dividen en subconjuntos basados en el valor de la característica seleccionada en el paso anterior. Este proceso se repite recursivamente para cada subconjunto creado en el paso anterior hasta que se cumple algún criterio de parada, como la profundidad máxima del árbol, el número mínimo de muestras en un nodo hoja, o si la pureza alcanza un umbral predefinido.


```python
display.Image("img/tree_based.png")
```

Los árboles de decisión son fáciles de usar e interpretar, y se pueden aplicar tanto a problemas de regresión como de clasificación. Sin embargo, tienen tendencia al sobreajuste (*overfitting*). Para mitigar esta situación, se pueden introducir técnicas de podado, que consisten en eliminar nodos del árbol que no aportan significativamente a la capacidad de generalización del modelo. Otra opción es usar conjuntos de árboles (*ensambles*), como los modelos de *random forest* o de *gradient boosting*.

Los modelos de conjuntos de árboles se basan en construir múltiples árboles de decisión durante el entrenamiento y combinar sus predicciones para obtener una predicción más robusta y precisa. Hay distintas maneras de seleccionar y combinar los árboles (generalmente con *boosting* o *bagging*) y ésta es una de las principales diferencias entre los distintos algoritmos. *Bagging* construye múltiples modelos independientes y los combina mediante votación o promedio, mientras que *Boosting* construye modelos secuencialmente, dando más peso a las instancias mal clasificadas en cada iteración.

```python
display.Image("img/boosting_bagging.png")
```

En el artículo anterior de pacientes con hepatitis, usando un modelo de *random forest* se mejoró la accuracy hasta el 85%. A parte de un mejor resultado, los modelos basados en árbol permiten obtener la contribución de cada variable en el modelo, resaltando la importancia de cada una.

### 3. Modelos de aprendizaje automático no supervisado <a name="id4"></a>

En el aprendizaje no supervisado, hay dos grandes grupos de técnicas: clustering y técnicas de reducción de dimensionalidad.

Técnicas como el análisis de componentes principales (*Principal component analysis, PCA*) o el algoritmo de t-SNE (*t-Distributed Stochastic Neighbor Embedding*) se usan para reducir el número de variables en el dataset, ya sea para evitar problemas de dimensionalidad de los datos o para ayudar en la visualización. Respecto a la reducción de dimensionalidad, estas técnicas se usan en el ámbito biomédico sobretodo para intentar mitigar el problema de la discrepancia entre un volumen alto de características en comparación con el reducido número de muestras (e.g. pacientes). Por otro lado, el algoritmo de t-SNE es muy usado actualmente en el ámbito de la transcriptómica, para poder visualizar los distintos grupos de expresión genética.

```python
display.Image("img/t-SNE_visualization_transcriptomics.png")
```

Los algoritmos de clustering tienen como objetivo dividir un conjunto de datos en grupos de manera que los objetos dentro de cada grupo sean más similares entre sí que con los objetos de otros grupos. Estos grupos se forman de acuerdo con algún criterio de similitud, como la distancia euclidiana o la similitud coseno. Algunos de los algoritmos más comunes son *K-Means*, *DBSCAN*, *Hierarchical Clustering* y *Gaussian Mixture Models* (GMM), y su elección depende de los datos y el problema que se quiere abordar.

Los modelos de clustering tienen muchas aplicaciones, especialmente en relación a la identificación de subgrupos relevantes entre poblaciones de pacientes. El hecho de no necesitar una clasificación previa permite descubrir patrones significativos, que después pueden relacionarse tambén con un tratamiento o una evolución diferente.

En el artículo `Phenotype clustering in health care: A narrative review for clinicians` se destaca el valor que pueden aportar los algoritmos de clustering en sacar información valuosa de datos biomédicos de distintas fuentes y tamaños.

### 4. Caso práctico: Aprendizaje automático aplicado a la diabetes y enfermedades cardiovasculares <a name="id2"></a>

**Artículo**: *Machine learning in precision diabetes care
and cardiovascular risk prediction*

**Autores**: Evangelos K. Oikonomou, Rohan Khera. Yale School of Medicine, New Haven, CT, USA.

La diabetes es una enfermedad que afecta a un gran nombre de personas en el mundo (almenos 529 millones) y está relacionada con un mayor riesgo de patologías cardiovasculares. Debido a la prevalencia significativa en la población, numerosos grupos de investigación y clínica se han centrado, y se centran, en encontrar herramientas para mejorar el diagnostico, seguimiento y tratamiento de estos pacientes.

Las técnicas de aprendizaje automático han sido también una revolución en el ámbito médico, y des de su auge, múltiples investigaciones se han centrado en aplicar estos modelos en el diagnóstico y tratamiento personalizado de la diabetes.

Sin embargo, aunque los resultados de los estudios son positivos en muchos casos, no se tienden a traducir en una mejora en la práctica clínica. Además, la necesidad de los algoritmos supervisados de tener datos etiquetados, perpetúa los sesgos del sistema actual. Así pues, es importante conocer las limitaciones de estas técnicas y usarlas con cuidado, para evitar hacer un mal uso.

Este artículo revisa diferentes métodos de inteligencia artificial aplicados al ámbito de la diabetes y se identifican tanto los puntos prometedores como las mejoras necesarias para que haya una aplicación clínica real de estas metodologías.

```python
display.Image("img/diab_review.png")
```

Algunas aplicaciones de métodos de ML en diabetes:

**Predicción del riesgo de diabetes con individuos preclínicos**

Actualmente se usan cuestionarios que evalúan determinados factores de riesgo para dignosticar adultos sin sintomatología. Con esta idea en mente, diversos estudios se han centrado en usar diferentes tipos de características para predecir el riesgo de padecer diabetes. Los resultados son positivos en la mayoría de ejemplos, con valores ed AUROC de más de 0.8, y valores de precisión/sensibilidad/especificidad altos. Por otro lado, diferentes algoritmos se han usado, como regresión logística y XGBoost.

**Predicción del riesgo de complicaciones cardiovasculares**

Los pacientes con diabetes tienen más riesgo de padecer enfermedades cardiovasculares. Para poder detectar estos problemas, se requieren técnicas no invasivas de imagen o pruebas funcionales, que suelen ser caras y dificiles de aplicar para hacer un *screening* de pacientes. Por eso, se han usado también modelos supervisados para predecir el riesgo de enfermedad cardiovascular. Los resultados son también positivos, y permiten también identificar qué factores son más influyentes en la detección de estas complicaciones cardiovasculares.

**Detección de distintos fenotipos en pacientes de diabetes**

Aunque tradicionalmente la diabetes se divide en dos tipos, dentro de un mismo grupo, los pacientes pueden presentar características diversas, que pueden dar lugar a manifestaciones clínicas heterogéneas. De ahí la necesidad de identificar subgrupos dentro del mismo tipo, para poder tener un enfoque más personalizado. En este punto, las técnicas no supervisadas pueden ser muy útiles. Diferentes algoritmos no supervisados (clustering, kmeans, NLP) se han aplicado a distintos tipos de datos (datos clínicos, genómicos, imágenes) para identificar grupos (clusters) de pacientes con características clínicas y prognosis distintos.


**Puntos importantes a tener en cuenta**

- El diseño del estudio es esencial para evitar problemas de overfitting en los resultados, que impactan la generalización e interpretabilidad de los modelos. Por tanto, es necesario implementar una buena separación en muestras de entrenamiento, validación y testeo, y añadir validación cruzada.

- La interpretabilidad de los resultados es clave. Por lo tanto, modelos muy complejos no siempre serán la mejor opción, sobretodo si implican una mayor dificultad en la interpretación de los resultados.

- Es importante pues tener en cuenta los posibles sesgos de los datos, debido a desigualdades en el acceso a la sanidad o al nivel socioeconómico de los pacientes.
