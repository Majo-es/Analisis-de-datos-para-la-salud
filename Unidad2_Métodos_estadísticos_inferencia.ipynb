Para verlo de manera más práctica, vamos a simular un posible dataset con datos de salud. En este caso, el dataset incluirá, para cada paciente, información sobre su edad, género, colesterol total y presión arterial sistólica (PAS).

Con estos datos vamos a realizar un análisis descriptivo simple.

```python
# Cargamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos un conjunto de datos de ejemplo
data = {
    'Edad': np.random.randint(18, 80, size=100),
    'Género': np.random.choice(['Mujer', 'Hombre'], size=100),
    'Colesterol_Total': np.random.randint(120, 250, size=100)
}

df = pd.DataFrame(data)

# Vamos a crear la variable Presión arterial sistólica (PAS)
# a partir de Colesterol_Total para que estén relacionadas

# Especificamos la desviación estándar del ruido
std_dev = 5
# Creamos un ruido de distribución normal
noise = np.random.normal(0, std_dev, len(df))

# Añadimos ruido a la variable 'Colesterol_Total'
df['PAS'] = df['Colesterol_Total'] + noise

# Hacemos un resumen estadístico con describe()
summary_stats = df.describe()

print("Resumen Estadístico:")
print(summary_stats)

# Creamos los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Creamos un histograma de edades
axs[0, 0].hist(df['Edad'], bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Distribución de edades')
axs[0, 0].set_xlabel('Edad')
axs[0, 0].set_ylabel('Frecuencia')

# Creamos un boxplot del colesterol total
axs[0, 1].boxplot(df['Colesterol_Total'])
axs[0, 1].set_title('Boxplot de colesterol total')
axs[0, 1].set_ylabel('Colesterol total (mg/dL)')

# Creamos un boxplot de PAS
axs[1, 0].boxplot(df['PAS'])
axs[1, 0].set_title('Boxplot de Presión Arterial Sistólica (PAS)')
axs[1, 0].set_ylabel('PAS (mmHg)')

# Creamos un scatter plot de edad vs. colesterol total
axs[1, 1].scatter(df['PAS'], df['Colesterol_Total'], color='coral', alpha=0.7)
axs[1, 1].set_title('PAS vs. Colesterol total')
axs[1, 1].set_xlabel('PAS')
axs[1, 1].set_ylabel('Colesterol total (mg/dL)')

# Mostramos la figura
plt.show()


```

**Análisis Inferencial**

Las herramientas explicadas en el apartado anterior son útiles para llevar a cabo un análisis exploratorio. Sin embargo, en la mayoría de los casos, el interés de un estudio estadístico va más allá de la caracterización descriptiva de cada variable y, en lugar de eso, el objetivo principal es obtener información y conclusiones sobre las características de estas variables en los grupos analizados. Este concepto se conoce como **inferencia estadística**.

La estadística inferencial se usa para realizar inferencias o generalizaciones sobre una población más amplia basándose en una muestra de datos.

En el ámbito de la salud, la estadística inferencial puede aplicarse para realizar afirmaciones sobre la eficacia de un tratamiento, comparar grupos de pacientes, y evaluar la asociación entre variables, entre otros. Por ejemplo, se puede realizar una prueba de hipótesis para determinar si un nuevo medicamento tiene un impacto significativamente diferente en la mejora de la salud en comparación con un placebo.

Algunas técnicas comunes en estadística inferencial incluyen:

- **Pruebas de Hipótesis:** Determinar si hay evidencia suficiente para rechazar o no una afirmación sobre una población.
- **Regresión:** Evaluar la relación entre variables y predecir valores futuros.

**Pruebas de hipótesis**


Hay diferentes tipos de **pruebas de hipótesis**. Un ejemplo común es examinar si un factor, que define a dos o más grupos, provoca diferencias en los valores poblacionales (como la media, la varianza o la proporción) de una variable específica (la característica de interés). Para poder comprobar esta hipótesis, se pueden usar los tests de comparación de medias y el análisis de la varianza (ANOVA) para la comparación de dos o más medias, respectivamente. También se incluyen los tests de comparación de varianzas para variables numéricas y los contrastes de homogeneidad para comparar proporciones en el caso de variables categóricas.

Este tipo de métodos pueden usarse para comparar, por ejemplo, de la media del número de linfocitos, o el porcentaje de supervivencia a los 5 años después de un diagnóstico, entre pacientes que han recibido diferentes tratamientos médicos.

Algunos de estos estudios pueden abordarse mediante dos categorías de técnicas estadísticas: **pruebas paramétricas** y **pruebas no paramétricas**. Las pruebas paramétricas suponen que las variables en las poblaciones que se están comparando siguen distribuciones estadísticas específicas, por ejemplo, una distribución normal. Por lo tanto, para garantizar la fiabilidad de las conclusiones del estudio, es necesario que se valide que los datos se ajustan a esa distribución. No obstante, varias de estas pruebas paramétricas, como los tests de comparación de medias o el ANOVA, pueden aplicarse incluso a datos que no sigan una distribución normal, siempre y cuando se identifiquen y excluyan los datos atípicos, y se utilicen tamaños de muestra considerables en cada población comparada (por ejemplo, >30).

En cambio, las pruebas no paramétricas no presuponen que los datos sigan una distribución estadística específica. Sin embargo, esto no implica que no estén sujetas a ciertas restricciones, como la similitud en las varianzas entre los grupos que se comparan. Aunque las pruebas paramétricas suelen tener una mayor potencia estadística, lo que aumenta la probabilidad de detectar un efecto significativo cuando existe, las pruebas no paramétricas son generalmente más robustas frente a la presencia de datos atípicos.

Imaginad que participamos en un estudio clínico que mide la eficacia de un fármaco para controlar los niveles de colesterol en pacientes con diabetes.

En este caso, tenemos dos grupos, unos pacientes que han tomado el fármaco que queremos evaluar, y otros pacientes que han tomado un placebo. El placebo es un producto o tratamiento simulado que no tiene efectos farmacológicos sobre la condición médica estudiada. Se usa como un control en ensayos clínicos para evaluar la eficacia de un tratamiento real y para separar los efectos reales de un tratamiento de los efectos psicológicos o subjetivos que pueden surgir debido a las expectativas del paciente o del personal médico.

La hipótesis del estudio es que este nuevo fármaco es eficaz para reducir el colesterol. Para comprobar la veracidad de esta afirmación, tenemos que realizar un análisis de inferencia. Dado que tenemos dos grupos y queremos evaluar si hay una diferencia significativa en una variable numérica, tenemos que realizar un **t-test** o una prueba **U de Mann-Whitney**. Para saber si tenemos que usar una prueba paramétrica o no paramétrica, se tiene que realizar previamente un test de normalidad, por ejemplo, la prueba de **Shapiro-Wilk**.

```python

## Situación 1

# Cargamos las librerías necesarias
import scipy.stats as stats

# Creamos dos grupos de pacientes
grupo1 = np.random.normal(loc=180, scale=20, size=50)
grupo2 = np.random.normal(loc=175, scale=18, size=50)

# Consideramos los dos grupos juntos

# Prueba de Shapiro-Wilk
stat, p_value = stats.shapiro([grupo1, grupo2])

# Imprimir el resultado
print(f"Estadístico de prueba: {stat}")
print(f"P-valor: {p_value}")

# Interpretar el resultado
if p_value > 0.05:
    print("Los datos parecen seguir una distribución normal.")
else:
    print("Los datos no siguen una distribución normal.")
```

```python
# Como los datos son normales, podemos usar un test paramétrico
# Calculamos un t-test de dos muestras
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)

# Imprimimos los resultados
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretamos el resultado
if p_value < 0.05:
    print("Hay una diferencia significativa entre los grupos.")
else:
    print("No hay evidencia suficiente para rechazar la hipótesis nula.")
```

```python

## Situación 2

# Cargamos las librerías necesarias
import scipy.stats as stats

# Creamos dos grupos de pacientes
grupo1 = np.random.randint(100, 250, size=50)
grupo2 = np.random.randint(100, 250, size=50)

# Consideramos los dos grupos juntos

# Prueba de Shapiro-Wilk
stat, p_value = stats.shapiro([grupo1, grupo2])

# Imprimir el resultado
print(f"Estadístico de prueba: {stat}")
print(f"P-valor: {p_value}")

# Interpretar el resultado
if p_value > 0.05:
    print("Los datos parecen seguir una distribución normal.")
else:
    print("Los datos no siguen una distribución normal.")
```

```python
# Como los datos no son normales, debemos usar un test no paramétrico
# Calculamos la U de Mann-Whitney
t_stat, p_value = stats.mannwhitneyu(grupo1, grupo2)

# Imprimimos los resultados
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretamos el resultado
if p_value < 0.05:
    print("Hay una diferencia significativa entre los grupos.")
else:
    print("No hay evidencia suficiente para rechazar la hipótesis nula.")
```

**Análisis de regresión**

El análisis de regresión es una técnica estadística utilizada para examinar la relación entre una o más variables independientes (predictoras) y una variable dependiente (respuesta). Su objetivo principal es comprender cómo las variables independientes afectan a la variable dependiente y predecir el valor de esta última en función de las primeras.

Primero de todo, se tiene que definir el problema, es decir, qué problema queremos abordar. Por ejemplo, ¿cómo afecta el nivel de colesterol en la sangre a la presión arterial sistólica?

Una vez se tienen el dataset listo, se tiene que seleccionar el modelo de regresión en función del tipo de datos y la relación esperada entre las variables. Por ejemplo, existen diferentes tipos de modelos de regresión, como la regresión lineal simple, la regresión lineal múltiple, la regresión polinomial, entre otros.

A continuación se ajusta el modelo de regresión a los datos mediante un método de estimación, como el método de mínimos cuadrados ordinarios. Este proceso implica encontrar los coeficientes que mejor se ajusten a los datos observados.

Para entender si la relación entre las variables es significativa, se evalua la calidad del ajuste del modelo utilizando diversas métricas, como el coeficiente de determinación (R-cuadrado), que indica la proporción de la variabilidad de la variable dependiente que es explicada por el modelo. También se realizan pruebas de hipótesis para determinar si los coeficientes del modelo son estadísticamente significativos. Esto ayuda a determinar si las variables independientes tienen un efecto significativo sobre la variable dependiente.

Finalmente, una vez que se ha validado el modelo, se puede utilizar para hacer predicciones sobre el valor de la variable dependiente para nuevas observaciones. Además, se pueden interpretar los coeficientes del modelo para comprender la magnitud y la dirección de la relación entre las variables.



En Python, el proceso de ajustar y evaluar el modelo se puede hacer usando la librería `statsmodels`. A continuación veremos como ajustar un modelo lineal de datos de colesterol y presión arterial simulados mediante OLS. *Ordinary Least Squares* es un método de estimación muy usado para encontrar los coeficientes que mejor ajustan un modelo a los datos observados.

La librería de statsmodels no solo realizar el ajuste, sino que también muestra un resumen de los resultados del modelo, que incluye estadísticas como el coeficiente de regresión y el valor p.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Generamos datos de ejemplo
np.random.seed(0)
# Nivel de colesterol
colesterol = np.random.normal(200, 30, 100)
# Presión arterial sistólica
PAS = 100 + 0.5 * colesterol + np.random.normal(0, 10, 100)

# Creamos un DataFrame
data = pd.DataFrame({'colesterol': colesterol, 'PAS': PAS})

# Añadimos una constante a los datos
X = sm.add_constant(data['colesterol'])

# Ajustamos el modelo de regresión lineal
modelo = sm.OLS(data['PAS'], X).fit()

# Imprimimos los resultados del modelo
print(modelo.summary())
```

### 3. Análisis de supervivencia <a name="id3"></a>

El análisis de supervivencia es una técnica estadística usada para estudiar el tiempo que transcurre hasta que ocurre un evento específico de interés. Este tipo de análisis se usa mucho en estudios e investigación médica, pero también se puede aplicar en otros campos, como la ingeniería, la economía y la sociología.

La variable de interés en el análisis de supervivencia es el tiempo hasta que ocurre un evento, y este evento puede ser de diversos tipos, como la muerte, la recaída de una enfermedad, la ocurrencia de un fallo en un sistema, etc.

Este enfoque es valioso en la investigación clínica para entender la eficacia de tratamientos, evaluar la progresión de enfermedades y hacer predicciones sobre el tiempo hasta eventos importantes en salud (por ejemplo, recaída de una enfermedad o fallecimiento).

En este contexto, el análisis de supervivencia se puede usar para responder a preguntas como:
- ¿Cuántos pacientes sobrevivirán más allá de un determinado tiempo?
- Los pacientes que sobreviven, ¿a qué ritmo morirán?
- ¿Cómo ciertas variables (edad, sexo, tipo de tratamiento, etc.) aumentan o disminuyen la probabilidad de supervivencia?

En el ámbito médico, el análisis de supervivencia tiene como objetivo clave estimar la **función de supervivencia** (S(t)), que representa la probabilidad de que un individuo o un elemento sobreviva más allá de un tiempo específico "t". Cuando esta función se representa gráficamente se conoce como **curva de supervivencia**.

Para definir la función de supervivencia, se pueden usar distintos modelos:
- Modelos Exponenciales (modelo paramétrico)
- Modelos de Kaplan-Meier (modelo no paramétrico)
- Modelos de Regresión de Cox (modelo semi-paramétrico)


Cuando se realiza el análisis de supervivencia para examinar las curvas de supervivencia de los sujetos, el enfoque más común es optar por un modelo de Kaplan-Meier. En caso de querer comprender qué características de un sujeto tienen roles importante en la determinación de la supervivencia, se puede usar un modelo de regresión de Cox.

Como cada paciente es diferente, es posible que el evento que se quiere detectar (por ejemplo, la recaída de una enfermedad) no se observe por algunos individuos dentro del período de estudio, lo que da lugar a **observaciones censuradas**. La censura puede surgir de las siguientes formas, por ejemplo paciente aún no ha experimentado el evento, o el paciente no puede o no quiere continuar el seguimiento.

Vamos a trabajar estos conceptos en la parte práctica de la sesión.
